### **Variables**

```python
M = 9000
K = 10
L = 10.5

plt.rcParams["figure.figsize"] = (20, 10)
```

```python
first_distribution = np.random.uniform(0.0, 1.0, M)
```

#### **Plot distribution of this vector of random numbers**

```python
count, bins, ignored = plt.hist(first_distribution,100, density=True)
plt.plot(np.linspace(0, 1, 1000), np.full(1000, 1))
plt.show()
```

#### **Show it's first ten elements and size as a proof**

```python
print(f"Ten elements of distribution are:\n {first_distribution[100:200:10]} \n And is't full size is {first_distribution.size}")
print(f"Max element in second distribution is: {first_distribution.max()} and minimal is: {first_distribution.min()}") 
print(f"Mean of this distribution is: {first_distribution.mean()}")
```

```python
second_distribution = np.random.poisson(L, M)
```

```python
count, bins, ignored = plt.hist(second_distribution,24, density=True)
plt.show()
```

```python
print(f"Fisrt ten elements of second distribution are:\n {second_distribution[:K]} \n And is't full size is {second_distribution.size}")
```

```python
print(f"Max element in second distribution is: {np.max(second_distribution)} and minimal is: {np.min(second_distribution)}") 
print(f"Mean of this distribution is: {second_distribution.mean()}")
```

```python
y_axis = np.array([second_distribution[:ind].mean() for ind in range(1, second_distribution.size)])
plt.plot(np.log(np.linspace(1, y_axis.size, y_axis.size)), y_axis, c="blue")
plt.title("The dependence of the mean value on the logarithm of the size")
```
