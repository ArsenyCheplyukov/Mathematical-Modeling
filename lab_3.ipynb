```python
M = 9000
generated_points = 1000
function_array = np.linspace(0, np.pi/2, generated_points)
generated_array = np.random.choice(size=M, a=function_array, p=np.sin(function_array)/np.sum(np.sin(function_array)))
```

```python
generated_array
```

```python
plt.hist(generated_array, 100, density=True);
plt.plot(np.linspace(0, np.pi/2, 1000), np.sin(np.linspace(0, np.pi/2, 1000)), color="red");
```

```python
def sin(x):
    return np.sin(x)
print(f"The mean value of this generated array is: {generated_array.mean():.2f} \nThe sum of numbers divided by it's length is: {(generated_array.sum())/M:.2f} \nThe theoretical mean value is: {integrate.quad(sin, 0, np.pi/2)[0]:.2f}")
```

```python
def x_sin(x):
    return np.power(x, 2)*np.sin(x)
print(f"The variance of this vector is: {generated_array.var():.2f}\nThe variance calculated by formula is: {np.power(generated_array, 2).sum()/(M-1) - np.power(generated_array.sum(), 2)/(M*(M-1)):.2f}")
print(f"The theoretical value is {integrate.quad(x_sin, 0, np.pi/2)[0]-np.power(integrate.quad(sin, 0, np.pi/2), 2)[0]:.2f}")
```
