```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import chi2

from numpy import random
from  scipy.stats import chi2_contingency
```

```python
plt.rcParams["figure.figsize"] = (20, 10)
```

```python
M = 9000
```

```python
def function(x):
    return np.power(x, -2)

def Neimans_generator(size):
    answer = np.array([])
    i = 0
    while(i < size):
        x = random.uniform(0.00001, 4, 1)
        y = random.uniform(0.00001, 4, 1)
        if y <= function(x):
            answer = np.append(answer, [x])
            i += 1
    return answer
```

```python
x_array_emperical = np.linspace(0.5, 1, M)
y_array_emperical = np.power(x_array_emperical, -2)
y_array_generator = Neimans_generator(M)
```

```python
plt.plot(x_array_emperical, y_array_generator, color="red");
plt.plot(x_array_emperical, y_array_emperical, color="green");
```

```python
kf = chi2_contingency(np.array([y_array_emperical, y_array_generator]))[1]
print(f"P-value is: {kf}\nIf p-value is greater than 0.05 it means that this two generation of values are not simmilar at all")
```
