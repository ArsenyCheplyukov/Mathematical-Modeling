```python
import numpy as np
from matplotlib import pyplot as plt

import plotly.graph_objects as go
from plotly.subplots import make_subplots

import seaborn as sns

import pandas as pd
from  scipy.stats import chi2_contingency
```

```python
plt.rcParams["figure.figsize"] = (20, 10)
```

```python
M = 22
x_size = 1000
generated_x = np.linspace(0, M, M)
p = ((generated_x+1)*np.power(0.7, generated_x) * 0.09)
p /= np.sum(p)
s = np.random.choice(a=generated_x, size=x_size, p=p)
print(f"The mean value of this generated array is: {s.mean():.2f}")
print(f"The theoretical value is: {s.sum()/(s.size):.2f}")
print(f"The standart deviation: {s.var():.2f}")
print(f"The variance calculated by formula is: {np.power(s, 2).sum()/(x_size-1) - np.power(s.sum(), 2)/(x_size*(x_size-1)):.2f}")
```

```python
plt.hist(s, 10, density=True);
plt.plot(np.linspace(0, M, x_size), ((np.linspace(0, M, x_size)+1)*np.power(0.7, np.linspace(0, M, x_size)) * 0.09), color="red");
```

```python
kf = chi2_contingency(np.array([s[:22], ((generated_x+1)*np.power(0.7, generated_x) * 0.09)[:22]]))[1]
print(f"P-value is: {kf}")
```
